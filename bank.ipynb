```python
data_encoded = data.copy()

# Rename 'deposit' to 'y' once
data_encoded = data_encoded.rename(columns={'deposit':'y'})

# Encode other object columns
for col in data_encoded.columns:
    if data_encoded[col].dtype == 'object' and col != 'y':
        data_encoded[col] = LabelEncoder().fit_transform(data_encoded[col])

# Encode target 'y'
data_encoded['y'] = data_encoded['y'].map({'yes':1, 'no':0})
```

```python
for col in categorical_features:
    contingency = pd.crosstab(data_encoded[col], data_encoded['y'])
    chi2, p, dof, ex = chi2_contingency(contingency)
    print(f"{col}: p-value = {p:.4f}")
```

```python
for col in numerical_features:
    yes = data_encoded[data_encoded['y']==1][col]
    no = data_encoded[data_encoded['y']==0][col]
    t, p = ttest_ind(yes, no)
    print(f"{col}: p-value = {p:.4f}")
```

```python
X = data_encoded.drop(['y'], axis=1)
y = data_encoded['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

print("Logistic Regression Results:")
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

print("Decision Tree Results:")
print(confusion_matrix(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print("Random Forest Results:")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
y_probs = rf.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label="Random Forest (AUC = %0.2f)" % roc_auc_score(y_test, y_probs))
plt.plot([0,1], [0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
